# 딥러닝 프로세스 (Deep Learning Process)   
   
1. 데이터 불러오기
2. 데이터 전처리(Data Preprocessing)

Data를 Model에 입력하기 전에 데이터를 '가공'하는 단계   
   
가지고 있는 데이터의 종류, 모델에 적용하려는 훈련 방법 등에 따라서 전처리 방법은 천차만별   
   
경우에 따라서는 배열의 차원에 변경이 이루어지기도 하고, 스케일의 조정이 이루어지기도 함   
   
데이터 전처리를 제대로 진행해야 다음에 진행할 모델 생성에서 올바르게 모델 정의를 할 수 있음   
   
3. 데이터 분할(학습용, 테스트용)   

4. 모델 생성   
   
모델의 구조를 파악하고 만들어내는 단계
- 모델 생성 방법   
  - Sequential API : 순차적 구조의 모델
  - Functional API : 다중 입력, 다중 출력을 가지고 있는 복잡한 구조의 모델

5. 모델 컴파일(Compile)   
만들어진 모델 훈련에 사용할 다양한 옵션 설정   
   
.compile() 함수를 사용
- lossfunction
- optimizer
- metrics
   
6. 모델 훈련   
   
.fit() 함수에 모델 훈련에 필요한 정보를 파라미터로 전달   
- 훈련 데이터셋
- 검증 데이터셋
- 크기(반복 횟수)
- 콜백(call-back)   
   
7. 모델 검증(evaluate)   
훈련이 끝난 모델이 얼마나 정확한지 검증하는 단계   
모델 훈련시 사용하지 않은 데이터셋을 입력시킨 후에, 그 모델이 예측한 값과 정답을 비교해서 평가지표를 산출함   
   
이 검증결과를 바탕으로 모델 생성단계로 돌아가서 컴파일, 모델 수정등이 이루어짐   
   
목표한 성능에 도달할 때까지 프로세스 반복   
   
8. 예측(Predict)   
   
임의의 데이터 or test용 데이터를 입력해서 그에대한 모델의 예측 결과값을 얻는 과정   
   
# 2,3,4,5 번의 경우에는 모델 학습을 위한 필수 프로세스로 절대 빼먹어서는 안됨!!

```python
(train_data, train_label), (test_data, test_label) = mnist.load_data()
```

```python
(train_data, train_label), (test_data, test_label)
```

```python
# 가져온 데이터에서 픽셀 최대값, 최소값 구하기
print(train_data.min())
print(train_data.max())
```

```python
# 이미지에 대한 내용을 정규화
train_data = train_data / train_data.max()
test_data = test_data / test_data.max()

print(test_data.shape)
# 값의 범위는 줄어들었지만, set의 형식 자체는 그대로
# (10000, 28, 28)

train_data[0, 5:10, 5:10]
# 정규화 후의 값은 0~1 사이의 값
```

```python
# 모델 생성
# Sequential() : 순차적 구조의 모델 >> 구조를 설정하고, 층(layer)을 설정하는 부분

# 대괄호 안쪽에 있는 것들이 각각의 층들임
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    # 1차원으로 평평하게 만들기 - 입력층
    # input_shape는 필수옵션
    # (28,28) 크기의 matrix가 들어갈테니 이걸 1차원 벡터로 변환해달라는 뜻

    tf.keras.layers.Dense(128, activation='relu'),
    # 은닉층(중간층)에 신경망을 만드는 단계
    # 128개짜리 perceptron으로 구성되는 layer 하나
    # activation='relu' : 은닉층에서 동작할 활성화 함수 중 하나

    tf.keras.layers.Dropout(0.2),
    # 데이터의 과대적합을 막아주는 역할
    # 128개의 신경망 중에 무작위 20%를 0으로 만들어서 비활성시키겠다.

    tf.keras.layers.Dense(10, activation='softmax')
    # 출력층
    # 10개짜리 perceptron으로 구성되는 layer 하나
    # activation='softmax' : 출력층에서 동작할 활성화 함수 중 하나
    # 10개로 구성한 이유?
    # 0~9 까지 총 10개의 이미지가 어떤 숫자를 의미하는지 파악하기 위해서
    # activation : 'softmax' : 출력층의 결과에 대한 다중 분류를 위해 softmax 사용(확률 값 형태로 분석되도록)

    # 활성화 함수(activation) : 결과값을 변환해서 다른 층으로 보내주는 함수
    # linear : 기본값, 입력 뉴런과 가중치로 계산된 결과값이 그대로 출력
    # relu : 은닉(중간) 층에서 사용
    # sigmoid : 이진분류 - 출력층 주로 사용
    # softmax : 다중분류 - 출력층에서 주로 사용(출력이 2개 넘으면 사용)
    ])
#요약
print(model.summary())
```

```python
# 컴파일
# .compile : 학습 방법에 대한 설정
# Sequential() 에서 정해진 모델을 컴퓨터가 알아들을 수 있는 말로 컴파일하는 부분

model.compile(
    optimizer='adam',
    # 최적화
    # 손실함수를 통해서 얻어낸 손실값으로부터 모델을 업데이트하는 방식

    loss='sparse_categorical_crossentropy',
    # 손실함수(loss function) : 모델을 최적화시킬 때 사용하는 함수
    # 손실함수는 신경망의 예측이 얼마나 잘 맞는지 측정하는 역할
    # categorical_crossentropy : 다중분류 손실함수 (one-hot encoding 클래스)
    # 출력값이 one-hot encoding 된 결과

    # sparse_categorical_crossentropy : 다중분류 손실함수 (one-hot encoding이 아님)
    # 출력값이 ineteger type 클래스로 변환

    # binary_crossentropy : 클래스가 2개인 이진 분류 손실함수
    # label이 0 or 1을 값으로 가질 때 사용

    # 정확도 확인
    metrics=['accuracy']


)
```

# 용어 정리

# 하이퍼파라미터(Hyper-Parameter)   
머신러닝이나 딥러닝에서 데이터를 훈련시킬 때, 조금 더 나은 조건에서 훈련할 수 있도록   
[사용자가 직접] 설정해주는 옵션   
   
학습 속도, 반복 횟수, 등은 사용자가 직접 설정 가능   
   
어떤 값을 어떻게 설정하는지에 따라서 그 모델의 성능이나 결과가 달라짐   
   
사용자가 따로 건들지 않으면 자동적으로 default값이 적용됨   
   
# 에포크 (epoch)   
반복 횟수를 지정   
데이터를 학습시키는 과정을 몇 번 반복해서 모델이 최적의 가중치를 찾아내게 할지 설정하는 방법   
   

# Underfitting(과소적합), Overfitting(과대적합)   
   
모델 학습에 있어서는 데이터는 크게 학습용 데이터 / 테스트(예측) 데이터로 구분   
   
반복적으로 학습을 시키면 모델은 사람이 발견하기 어려운 패턴을 발견하게 되어서   
사람의 예측보다 우월한 모델 생성이 가능   
   
예측용 데이터가 학습시킨 모델과는 다른 데이터의 분포를 가지고 있거나,   
학습시킨 데이터가 한쪽으로 치우친(편향된) 데이터라면...   
=> 그 모델은 예측 성능이 현저하게 떨어지게 됨...   
   
- Underfitting(과소적합) : 모델의 데이터 학습이 충분하지 않은 경우
=> 예측 성능 저하
- Overfitting(과대적합) : 학습 데이터를 지나치게 많이 반복해서 투입한 경우
=> 너무 편향된 모델이 될 가능성이 생김   
   
위의 두 문제를 피하면서 정확한 예측값을 내놓는 모델을 생성하는게 관건이다.

```python
# training, evaluation(훈련, 검증)
# .fit() : model을 실제로 학습시키는 단계
model.fit(train_data, train_label, epochs=10) #6만개의 데이터를 10번 반복

# 훈련 데이터에서 사용하지 않았던 (test_data, test_label)로 검증/평가
model.evaluate(test_data, test_label)
```
