# 토큰화 ( Tokenization )   
어떤 문장을 단어로 잘라내서 정제하고, 정규화를 시키는 과정   
   
구두점( Functuation ) 기반   
마침표, 쉼표, 물음표, 느낌표, 세미콜론 등등   
   


```python
text = 'Actions speak louder than words. A journey of a thousand miles begins with a single step. People who live in glass houses should not throw stones.'
#'A journey of a thousand miles begins with a single step.'
#'People who live in glass houses should not throw stones.'

```

```python
print(word_tokenize(text))
# 구두점 분리는 잘 됨
# Don't -> do n't으로 분리
# ain't -> ai n't으로 분리
# you're -> you 're 로 분리
print()

print(WordPunctTokenizer().tokenize(text))
# 구두점을 별도로 표시

print()

print(text_to_word_sequence(text))
# 모든 알파벳을 소문자로 치환
# 구두점 제거
# don't 를 통째로 단어처리
# 축약어를 통째로 단어처리(보존)
```

# 한국어 자연어 처리 : KoNLPy 라는 파이썬 패키지 사용   
   
KoNLPy에서 사용할 수 있는 형태소 분석기   
-Okt(Open Korean Text)      
-Komoran   
-kkma(꼬꼬마)   
-Mecab   
-Hannanum   

```python
pip install konlpy
```

```python
from konlpy.tag import Okt
from konlpy.tag import Kkma
```

```python
okt = Okt()

print(okt.morphs('오늘은 화요일이고요, 내일은 수요일입니다!'))
# morphs : 형태소 분석
# 어떤 대상의 어절을 최소 의미 단위인 형태소로 분석하는 것

print(okt.pos('오늘은 수요일이고요, 내일은 목요일입니다!'))
# pos : 품사 태깅

print(okt.nouns('오늘은 목요일이고요, 내일은 금요일입니다!'))
# nouns : 명사 추출
```

```python
kkm = Kkma()

print(kkm.morphs('오늘은 화요일이고요, 내일은 수요일입니다!'))

print(kkm.pos('오늘은 수요일이고요, 내일은 목요일입니다!'))

print(kkm.nouns('오늘은 목요일이고요, 내일은 금요일입니다!'))
```

# 코퍼스(Coupus) : 말뭉치   
보통 여러 단어들로 이루어진 문장, 분석하려는 대상, 문서 데이터셋 등등을 말함  
   
코퍼스에서 용도에 맞게 토큰을 나누는 것을 토큰화(Tokenization)라고 함   
   
토큰화를 진행하기 전, 후에 텍스트를 용도에 맞게 정제(Cleaning), 정규화(Normaliztation)가 필요함   
   
정제(Cleaning) : 코퍼스에서 노이즈 데이터(일반적으로는 조사, 접사 등등)를 제거하는 작업   
정규화(Normaliztion) : 표현 방법이 서로 다른 단어들을 통일시켜서 같은 단어로 재가공   
1. 규칙에 따라서 표기가 다른 언어를 통합시키기   
ex) US USA us U.S.A   
2. 대소문자를 통합
