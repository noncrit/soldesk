# 인공지능 ( Artifical intelligence )   
   
컴퓨터에게 데이터를 학습시켜서 특정 문제에 대한 판단이 가능하도록 하는 행위 : 기계학습   
   
AI : 인간의 지능과 유사한 지능을 가진 컴퓨터 시스템   
인간의 지능을 인공적으로 컴퓨터에게 심어놓은 것   
   
Machine Learning :   
학습 모형을 기반으로 외부에서 들어오는 데이터를 스스로 학습시키는 것   
빅데이터를 분석하거나, 가공해서 새로운 정보를 깨우치거나, 미래 예측이 가능하도록 하는 기술   
쌓아둔 데이터를 기반으로 특정 범주의 문제에 대한 상관 관계, 특성등을 찾는데 유리함  
   
Deep Learning :   
컴퓨터가 다양한 데이터를 인공 신경망 기반으로 스스로 학습이 가능하도록 모델을 구축한 형태   
특정 범주의 데이터에 특화된 것이 아닌, 기계 스스로의 학습 능력에 집중하는 형태
