```python
# 파이썬에서 정규 표현식을 사용할 때 내장 모듈인 re를 사용
import re
```

# 메타 문자로 표현해서 일대일 매칭   
   
# 메타 문자 : 특수한 기능을 하는 문자   
. : 한 개의 임의의 문자   
? : 앞 문자가 있을 수도, 없을 수도 있다(문자가 0개 또는 1개)   
\* : 별표가 붙은 문자가 0~임의의 갯수로 존재   
\+ : 앞의 문자가 최소 1개 이상 존재함   
\^ : 기호 뒤 문자열로 시작하는 문자열   
\$ : 기호 앞 문자열로 끝나는 문자열   
\{숫자} : 특정  숫자만큼 반복   
\[문자] : 대괄호 안의 문자들 중 한 개의 문자와 매칭함
\[^문자] : 해당 문자를 제외한 문자를 매칭함

```python
# re.match() 함수 : [문자열 처음]에 시작하는 패턴이 일치하는지 => 불일치시 None값 반환
# re.match(패턴, 문자열)
print(re.match('a', 'a'))
print(re.match('a', 'abc'))
print(re.match('a', 'bca'))
# 결과창의 span=(x,y) : matching되는 문자열의 위치를 반환, span(이상, 미만)
```

```python
# re.search() 함수 : 처음이 아니더라도 문자열 어딘가에 패턴에 해당하는 문자가 있으면 그 위치를 반환

print(re.search('a', 'a'))
print(re.search('a', 'bbb'))
print(re.search('a', 'bba'))
print(re.search('a', 'bqwefsdfwae'))
#
```

```python
# re.findall() 함수 : 문자열 중에서 패턴이랑 일치하는 모든 부분을 찾아줌
print(re.findall('a', 'a'))
print(re.findall('a', 'abca'))
print(re.findall('a', 'aaaa'))
print(re.findall('aaa', 'aaaa'))
#
```

```python
# re.finditer() : 패턴과 일치하는 모든 부분의 위치를 알 수 있음
where = re.finditer('a', 'abca')
for i in where:
    print(i)
```

```python
# re.fullmatch() : 패턴과 문자열이 완벽하게 일치할 경우에 그 위치를 반환
print(re.fullmatch('a', 'a'))
print(re.fullmatch('a', 'aaa'))
print(re.fullmatch('a', 'abcde'))
```

```python
matching = re.search('happy', 'Do not worry, be happy')
print(matching)

# match 관련 method
print(matching.group()) # group() : 일치하는 문자열을 반환
print(matching.start()) # start() : 일치하는 문자열의 시작 위치 반환
print(matching.end()) # end() : 일치하는 문자열의 끝 위치 반환
print(matching.span()) # span() : 일치하는 문자열의 위치를 튜플로 반환
#
```

# 대괄호 \[ ] : 여러가지 문자 중 하나와 일치   
[ ] 대괄호 안에 원하는 문자를 여러개 넣고, 문자열에서 대괄호에 넣은 문자 패턴과 일치하는게 하나라도 있으면 매칭   


```python
[a-z] : 소문자 a부터 z까지 중 하나와 일치
[A-z]
[A-Z]
[0-9]
#한글은?
[가-힣]
# 알파벳 대소문자 [A-Za-z]
```

```python
matching = re.fullmatch("You['-:~<>]re sleepy", "You're sleepy")
matching
```

```python
matching2 = re.search('happy [a-z]uesday', 'happy tuesday')
print(matching2)

matching3 = re.search("happy [^0-9]uesday", 'happy tuesday')
print(matching3)
```

```python
# re.compile() : 사용하고자 하는 패턴이 자주 이용될 때 미리 컴파일 하는 용도
# 마침표 . : 한 개의 임의의 문자
m = re.compile('a.c') # a,c 사이의 임의의 문자 1개가 있는 문자열
print(m.search('abc'))

m2 = re.compile('a..c')
print(m2.search('abbc'))
print(m2.search('asdfadfsdnibowfwebua'))
```

```python
# ? : 물음표 바로 앞에 있는 문자가 있을 수도, 없을 수도
q = re.compile('ab?c')

print(q.search('abbc'))
print(q.search('abc'))
print(q.search('abdc'))
#
```

```python
# * : * 바로 앞에 있는 문자가 0 개 이상일 때
s = re.compile('ab*c')

print(s.search('abbc'))
print(s.search('abc'))
print(s.search('abdc'))
#
```

```python
# + : *의 기능이랑 비슷하지만, + 앞에는 문자가 최소 한 개 이상
p = re.compile('ab+c')

print(p.search('abbc'))
print(p.search('abc'))
print(p.search('abdc'))
#
```

```python
# ^ : ^뒤에 나오는 문자열과 동일해야 함, 그 문자열로 시작하기만 하면 매칭이 가능
c = re.compile('^ab')

print(c.search('abbc'))
print(c.search('abc'))
print(c.search('babdc'))
#
```

```python
# {숫자} : {숫자} 바로 앞의 문자가 해당하는 숫자만큼 반복되어야 매칭 가능
n = re.compile('ab{3}c')

print(n.search('abbc'))
print(n.search('abbbc'))
print(n.search('abbbbc'))
print(n.search('abbbbbbc'))
#
```

```python
# {숫자, 숫자} : {숫자, 숫자} 바로 앞의 문자가 해당하는 숫자만큼 반복, 숫자는 이상, 이하를 나타냄
n2 = re.compile('ab{2,4}c')

print(n2.search('abbc'))
print(n2.search('abbbc'))
print(n2.search('abbbbc'))
print(n2.search('abbbbbbc'))
```

```python
# {숫자, } : 해당하는 숫자 이상이기만 하면 매칭됨
nnn = re.compile('ab{2,}c')

print(nnn.search('abbc'))
print(nnn.search('abbbc'))
print(nnn.search('abbbbc'))
print(nnn.search('abbbbbbc'))
```
