```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
#[['twinkle', 'twinkle', 'little', 'star'], ['wonder'], ['world', 'high'], ['like', 'diamond', 'sky'], ['blazing', 'sun', 'gone'], ['nothing', 'shines', 'upon'], ['show', 'little', 'light'], ['twinkle', 'twinkle', 'night'], ["trav'ller", 'dark'], ['thanks', 'tiny', 'spark'], ['could', 'see', 'way'], ['twinkle'], ['dark', 'blue', 'sky', 'keep'], ['often', 'thro', 'curtains', 'peep'], ['never', 'shut', 'eye'], ['till', 'sun', 'sky'], ['bright', 'tiny', 'spark'], ['lights', "trav'ller", 'dark'], ['tho', 'know'], ['twinkle', 'twinkle', 'little', 'star']]
final_sentence = [['twinkle', 'twinkle', 'little', 'star'], ['wonder'], ['world', 'high'], ['like', 'diamond', 'sky'], ['blazing', 'sun', 'gone'], ['nothing', 'shines', 'upon'], ['show', 'little', 'light'], ['twinkle', 'twinkle', 'night'], ["trav'ller", 'dark'], ['thanks', 'tiny', 'spark'], ['could', 'see', 'way'], ['twinkle'], ['dark', 'blue', 'sky', 'keep'], ['often', 'thro', 'curtains', 'peep'], ['never', 'shut', 'eye'], ['till', 'sun', 'sky'], ['bright', 'tiny', 'spark'], ['lights', "trav'ller", 'dark'], ['tho', 'know'], ['twinkle', 'twinkle', 'little', 'star']]
print(final_sentence)
```

```python
tokenizer = Tokenizer()

# fit_on_texts(코퍼스) : 빈도수를 기준으로 단어 집합을 만들어 줌
# 높은 빈도수의 단어부터 낮은 숫자를 부여 => 정수 인코딩
tokenizer.fit_on_texts(final_sentence)
```

```python
# word_index : 각각 어떤 인덱스가 부여되었는지 확인
print(tokenizer.word_index)
```

```python
# word_counts : 단어들의 빈도수 확인
print(tokenizer.word_counts)
```

```python
# texts_to_sequences(corpus) : text를 부여받은 인덱스로 변환하기
print(tokenizer.texts_to_sequences(final_sentence))
```

```python
# 빈도수가 높은 상위 단어들만 따로 집합을 구성하는 방식
# tokenizer = Tokenizer(num_words=숫자)
# 가장 빈번히 등장하는 숫자순으로 숫자값 만큼 토큰화할 단어 지정
word_size = 5
tokenizer = Tokenizer(num_words=word_size+1)
tokenizer.fit_on_texts(final_sentence)
print(tokenizer.word_index)
```

```python
# word_counts : 단어들의 빈도수 확인
print(tokenizer.word_counts)
```

```python
# texts_to_sequences(corpus) : text를 부여받은 인덱스로 변환하기
print(tokenizer.texts_to_sequences(final_sentence))
```

```python
# keras의 tokenizer 에 단어 집합에 없는 단어인 경우
# OOV에 대해서는 단어 자체를 제거하는 특성

freSize =5
tokenizer = Tokenizer(num_words=freSize+2, oov_token='OOV')
tokenizer.fit_on_texts(final_sentence)
print('단어 OOV의 인덱스 : {}'.format(tokenizer.word_index['OOV']))
```

```python
# oov_token을 사용하면 oov의 인덱스가 1로 설정
# 나머지 빈도수 상위 5개(단어 집합) 자동적으로 인덱스가 1씩 추가됨
print(tokenizer.texts_to_sequences(final_sentence))
```

```python
# 형태소 분석용
from konlpy.tag import Okt
okt = Okt()
```

```python
#불용어 집합 처리
sw = set(stop_words.split(' '))
# 형태소 분석
word_tokens = okt.morphs(text)
```

```python
# word_tokens에서 줄바꿈,특수 기호( 괄호 등등) 제거(regex를 이용해 여러 개 까지)
# re.sub -> replace 기능을 함
import re
word_tokens = [re.sub(r'\n', '', word) for word in word_tokens]
word_tokens = [re.sub(r'\(', '', word) for word in word_tokens]
word_tokens = [re.sub(r'\)', '', word) for word in word_tokens]

print(word_tokens)
```

```python
# 불용어에 포함된 단어는 모두 제거, 단어의 글자수가 1을 초과하는 단어만 포함시키기
# 그 단어들의 빈도수까지 확인 - dict 로 담기
# ex) {'기아':1....}
result = [word for word in word_tokens if not word in sw]

word_counts = {}
for word in result:
    if len(word) >1:
        if word not in word_counts.keys():
            word_counts[word] = 1
        else:
            word_counts[word] += 1
print(word_counts)
```

```python
# 빈도수를 기반으로 내림차순 정렬
word_counts = sorted(word_counts.items(), key = lambda x : x[1], reverse=True)
print(word_counts)
```

```python
# 빈도수가 높을수록 낮은 정수값을 부여
aa_index = {}
counter = 0
for (word, freq) in word_counts:
    if freq >1:
        counter +=1
        if len(word) >1:
            aa_index[word] = counter
print(aa_index)
```
