# 딥러닝 (Deep Learning)   
   
머신러닝의 한 종류   
   
여러 층의 신경망을 이용해서 머신 러닝을 수행하는 것   
   
보통 머신러닝과의 차이점 : [특징량 추출]   
   
이전에는 모델에게 사람들이 특징에 대한 데이터, 목적으로 하는 출력들을 학습시켜야 했음   
   
딥러닝을 이용해서 컴퓨터가 자동적으로 특징을 [스스로] 학습하도록!   


# 신경망 (Neural Network)   
   
사람의 신경망을 본따서 만든 네트워크 구조를 의미   
컴퓨터가 학습 능력을 갖게 해서 다양한 문제를 해결할 수 있게 하는 접근 방법   
   
신경망은 다양한 뉴런이 연결되어 있는 구조를 가지고 있는 네트워크 망

# 퍼셉트론 (Perceptron)   
   
인공 신경망으로 다수의 입력으로부터 하나의 결과를 내보내는 알고리즘   
   
비교적으로 구조가 단순하고, 현재 기계학습의 기초가 되는 개념   
   
입력층, 출력층으로 구성된 단순 퍼셉트론   
   
ex) 옷을 구매한다고 할 때   
보통 생각하는 조건들   
- 1 : 이번달 수입에 여유가 있는지   
- 2 : 지금 옷장에 옷이 없는지   
- 3 : 지금 당장 옷이 필요한지(옷에 문제가 있는지)   
   
사람에 따라, 경우에 따라 3가지 조건 중 중요하다고 생각하는 비율이 다름   
   
퍼셉트론 : 각 입력에 대한 [가중치] 도입   
   
다양한 조건 및 가중치를 기반으로 구입할지, 말지의 여부를 정하는 것   
-> 역치(선택의 기준이 되는 값)   
   
가중치와 역치를 고려해서 의사 결정 상황에 따라 반영이 가능
