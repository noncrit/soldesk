# DecisionTree(의사결정 나무) : 가지치기   
   
Superivsed Learning - classification   
입력값에 대한 예측값을 나무 형태로 나타내주는 모델   
데이터를 분류하거나 연속된 값을 예측하기 위해서 사용하는   
Tree 구조의 모델   
   
트리 구조는 사람이 이해하기 쉽고, 각 노드의 조건이 어떤 결정을 내리는지 명확하게 보여줌   
   
계층적인 구조(트리 형태)로 의사 결정을 내리며,   
각 단계에서 데이터를 특정 기준에 따라 나누는 과정을 통해서 최종 결과를 도출 => 데이터들의 분포 역시 나눠줌   
   
데이터를 어떤 기준으로 분류했을 때   
동일한 객체들로 잘 모아지게 할 수 있을까?   
=> 분류 기준을 잘 세우자   
   
각 노드에서 데이터를 분할하는 일련의 의사결정 과정을 통해 최종 결과를 도출   
   
- 특징
1. 알고리즘 해석이 쉬운 편
2. 수치형, 범주(카테고리(형 데이터도 적용 가능
3. 데이터 전처리 양이 적은 편



```python
from sklearn import datasets

iris = datasets.load_iris()
print(iris.feature_names) # feature (data, column, columns) 이름
print(iris.target_names) # target (label) 이름
```

```python
# scikit-learn이 제공하는 의사 결정나무 모델 : DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
feature = iris.data # feature 가져오기
label = iris.target # 목표 분류 라벨(target) 가져오기
# print(feature[:,2:]).head()
p_feature = feature[:,2:] # 붗꽃 꽃잎의 길이와 넓이

iris_tree = DecisionTreeClassifier(criterion = 'entropy', max_depth=3)
# DecisionTreeClassifier : 의사결정나무 모델
# criterion='entropy' : 엔트로피를 불순도 계산 방법으로 적용
# entropy : 혼잡도 , 학습 시작 단계에서는 분류할 것들이 많아서 엔트로피가 높은 상태
            # 하나씩 정리해가면서 엔트로피를 0에 가깝게 만드는 방식
# criterion = 'gini' (default) : 분류를 잘못할 확률을 최소화 하기 위한 목적
# -> 분할이 가능한 모든 방법에 대해 지니 계수를 확인 -> 계수가 가장 낮은 방식을 선택
# 지니계수 : 불확실성의 지표, 0에 가까울 수록 데이터의 불확실성이 낮아짐
# max_depth = 3 : 깊이 (몇 단계로 나눌지)
# 노드(Node) : 트리의 그 노드 맞음, 데이터의 분기, 결정이 이루어지는 부분
# 노드의 분류 - 루트(Root), 내부 노드(Internal Node), 리프(Leaf Node)(최종 지점)
# depth : 루트 - 리프 까지 도달하는 과정에서 거진 노드의 수
# 깊이를 제한하는 옵션으로 Tree가 지나치게 깊어지는 것을 방지
# 깊이를 제한하지 않으면 Overfitting 가능성이 커짐
# 깊이가 너무 얕으면 Underfitting 가능성이 커짐
# 깊이를 적절하게 제한하는게 상당히 중요 => 복잡도 줄이면서 오버피팅 방지

# 어떤 기준으로 분류를 하는가에 따라서 tree 모양도 변하고, 깊이도 달라짐

# 학습 시작
iris_tree.fit(p_feature, label)
# .fit(독립변수, 종속변수)
# 독립변수 : 원인, 학습할 데이터
# 종속변수 : 결과, 분류된 결과
# 종속변수는 독립변수에 의해 영향받음
```

```python
# dot 파일 만들기
from sklearn import tree
with open('irirs.dot', mode='w') as f:
  tree.export_graphviz(iris_tree, out_file=f)
# graphviz : 코드의 구조를 다이어그램 형태의 그림으로 표현해주는 도구
# out_file : 파일 or 문자열로 반환
```

# dot -Tpng [dot 파일명].dot -o [만들 png 파일명].png

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from subprocess import check_call

iris = datasets.load_iris()
x = iris.data[:,2:] #붗꽃 꽃잎의 길이, 넓이
y = iris.target

tree = DecisionTreeClassifier(max_depth=4)
tree.fit(x,y)

export_graphviz(
    tree,
    out_file='./iris2.dot',
    # 사용한 feature, columns
    feature_names = iris.feature_names[2:],
    #예측할  target class 이름
    class_names = iris.target_names,
    filled = True,  # 사각형(노드) 안에 색 채우기
    rounded = True  #사각형 모서리 둥글게
    )

# png 로 바꿔서 시작화
# dot -Tpng iris2.dot -o iris.png
# error handling 용 함수

check_call(['dot', '-Tpng', 'iris2.dot', '-o', 'iris.png'])

# dot : Graphviz 도구 중 하나인 dot 명령어, 그래프를 그림 형식으로 변환 시 사용
# -Tpng : 출력 파일의 형식을 지정하는 옵션
# -o : 출력 파일의 이름을 지정하기 위한 옵션
# iris.png : -o 옵션에 의해 지정된 출력 파일의 이름
```
