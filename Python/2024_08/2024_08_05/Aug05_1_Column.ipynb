# 2. 열 분리 : Series.str.get(인덱스)   


```python
flight = sns.load_dataset('flights')
flight.head()
```

```python
# 월에 영어값인거 => 각각 숫자로 변경해서 데이터에 적용
months = {
    'Jan' : 1,
    'Feb' : 2,
    'Mar' : 3,
    'Apr' : 4,
    'May' : 5,
    'Jun' : 6,
    'Jul' : 7,
    'Aug' : 8,
    'Sep' : 9,
    'Oct' : 10,
    'Nov' : 11,
    'Dec' : 12
}
flight['month'] = flight['month'].map(months)
flight.head()
```

```python
# month 컬럼 옆에 day 컬럼을 추가하고, 컬럼값은 모두 15로 생성
flight.insert(2, 'day', 15)
flight.head()
```

```python
# year, month, day 열을 합친 날짜데이터를 'date'라는 새로운 컬럼에 추가
flight['date'] = pd.to_datetime(flight.loc[:,['year','month','day']])
flight.head()
```

```python
#date 컬럼 다시 쪼개기
flight['date'] = flight['date'].astype('str')
dates = flight['date'].str.split('-')
dates.head()
```

```python
# dates 변수는 list[연, 월, 일]을 개별 요소로 갖는 시리즈 형태의 객체
# str.get() 함수를 이용하면 각 행별로 인덱싱이 가능
# str.get() : 리스트로 변환 후 해당 인덱스 위치의 요소를 반환
flight['y'] = dates.str.get(0)
flight['m'] = dates.str.get(1)
flight['d'] = dates.str.get(2)
flight.head()
```

# 2-1 매핑 함수를 이용해서 열 분리   
매핑함수 (람다함수)와 for문을 이용해서 위와 같은 결과 내기

```python
# 매핑함수 (람다함수)와 for문을 이용해서 date 쪼개기
flight['y'] = flight['date'].astype('str').map(lambda x : x.split('-')[0])
flight['m'] = flight['date'].astype('str').map(lambda x : x.split('-')[1])
flight['d'] = flight['date'].astype('str').map(lambda x : x.split('-')[2])
flight
```

```python
# 매핑함수 (람다함수)와 for문을 이용해서 date 쪼개기 ver. other
flight['date'] = flight['date'].astype('str')
col = ['y','m','d']
for i in range(len(col)):
  flight[col[i]] = flight['date'].map( lambda x : x.split('-')[i])
flight
```
