# 3-5. 상관 계수 ( 문자열은 제외 ) : corr()   
상관 관계 : -1 ~ 1사이의 값, 두 변수의 상관성에 대한 정보   
변수간 흐름이 얼마나 비슷한가   
a가 증가하면 b도 같이 증가 / 감소하는가   
증가의 성향이 얼마나 비슷한가 등등   
1에 가까울수록 두 지표는 비슷하게 증가 / 감소   
-1에 가까울수록 지표가 반대로 움직임   
0에 가까울 수록 지표가 관계가 없음   
   
값의 추이만 나타낼 뿐, 인과관계를 증명하지는 못함   
   
상관 계수는 독립변수와 종속변수로 만들어져야한다 (?)   
   
데이터 특징 분석시에는 요긴하게 쓸 수 있음 (EDA : Exploratory Data Analysis)   

corr() 함수는 문자열을 제외한 변수간 matrix를 생성하고, 각 쌍들의 상관계수를 반환   
Matrix - 각 요소의 자료형이 모두 일치해야함

```python
print(i.corr)
```
