```python
import pandas as pd
import seaborn as sns
```

```python
#엑셀 파일 넣은거 읽어오기
a= pd.read_excel('myStudent.xlsx')
df = pd.DataFrame(a)
df
```

```python
# kor, eng, math 를 더한 값을 'sum'열에 출력
# df['sum'] = df['kor'] + df['eng'] + df['math']
df['sum'] = df[['kor','eng', 'math']].sum(axis=1)
df['sum'].head()
```

```python
# 어떤 값을 넣으면 그 값을 3으로 나눈 결과를 내보내주는 함수
def div_by3(a):
  return a/3
print(div_by3(9))
```

```python
# 아까 만든 함수 매핑
df['avg'] = df['sum'].apply(div_by3).round(2)
df['avg'].head()
```

# 함수 매핑( Mapping )
   
      
시리즈나 데이터프레임의 개별 요소를 특정 함수에 1:1로 대응시키는 과정   
   
사용자가 직접 만든 함수(람다 함수 포함)를 적용시킬 수 있기 때문에,   
for문이나 기본 함수로 처리하는 것보다 효율적이다.

```python
# seaborn에서 타이타닉 데이터 불러와서 'age','fare' 열만 인덱싱
# => 'ten' 열을 만들어서 10을 넣기
titanic = sns.load_dataset('titanic')
df = titanic.loc[:,['age','fare']]
df['ten'] = 10
df.head()
```

```python
# 값을 넣으면 10을 더해주는 함수
def add_10(a):
  return a+10
# 값을 2개 넣으면 그 둘을 곱해주는 함수
def add_two_val(a,b):
  return a+b
# 값을 3개 넣으면 그 셋을 더해주는 함수
def add_three_val(a,b,c):
  return a+b+c
```

```python
# 위의 함수를 apply()를 이용해 시리즈 객체의 개별 요소에 적용
sr1 = df['age'].apply(add_10)
sr1.head()
```

```python
# apply 함수의 첫번째 파라미터는 적용시긴 함수, 나머지는 직접 지정하는 형태
sr2 = df['fare'].apply(add_two_val, b=0.9)
sr2.head()
```

```python

sr3 = df['fare'].apply(add_three_val, b=10.1, c=20)
sr3.head()
```

```python
# 'ten'열의 값들을 람다 함수를 통해 add_10 함수 적용
# (lambda a,b:print(a*b))(10,20)
sr4 = df['ten'].apply(lambda x: add_10(x))
sr4.head()
```

# 시리즈 객체 함수 매핑   
   
DataFrame에 apply 함수를 적용(매핑함수, axis=값)   
axis 를 통해 적용할 범위를 조정 가능( 행, 열 )

```python
#뭔가를 넣었을 때 결측값인지 확인하는 함수
def missing_value(series):
  return series.isnull()
```

```python
# 이 함수를 데이터프레임의 모든 열에 매핑( axis = 1 )
result = df.apply(missing_value, axis=1)
print(result)
print()
```
