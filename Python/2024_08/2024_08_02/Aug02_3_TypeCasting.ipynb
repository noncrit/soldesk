# 2. 범주형( 카테고리 ) 데이터 정리   
   
# 구간 분할 : np.histogram(), pd.cut() 활용

```python
aa = pd.DataFrame(a)
aa.head()
```

```python
# np.histogram() 함수 : np.histogram( 나눌 리스트, bins = 나눌 구간)
# return - 두 개의 객체가 리턴으로 돌아옴
#   count : 각 구간에 속하는 값의 갯수
#   bins_dividers : 각 구간의 경계값 리스트
count, bins_dividers = np.histogram(aa['avg'], bins=3)
print(count, '\n', bins_dividers)
```

```python
# pd.cut() : 자동으로 범주형 변수를 생성 - 간단함
bin_names = ['low','middle','high']
aa['aa_bin'] = pd.cut(
    x = aa['avg'], #대상 데이터 배열
    bins = bins_dividers, #경계값 리스트
    labels = bin_names, #구간명
    include_lowest = True # 첫 경계값 포함 여부(구간의 하위 값)
)
aa.head()
```
