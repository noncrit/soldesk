```python
import pandas as pd
```

# 1. csv 파일 읽고 쓰기

# 1-1. 쓰기 (to_csv)

데이터프레임을 만들고 csv 파일에 저장

```python
df = pd.DataFrame({'A':[0,1,2], 'B':[1,2,3], 'C':[4,5,6], 'D':[7,8,9]})
df.to_csv('test.csv')
df

# index = False 옵션은 데이터프레임에 자도응로 추가된 인덱스 없이 저장하겠다는 선언
# 옵션 지정 안한경우, rangeindex (0 부터 점점 증가하는 형식)가 디폴트로 들어감
```

1-2 읽기(read_csv)

```python
df1 = pd.read_csv('test.csv')
df1
# read_csv() 함수는 첫번째 행을 header(열이름)로 지정해서 불러옴
# 불러올 데이터에 미리 설정한 header가 없다면 header = None 옵션을 지정해줘야함

df2 = pd.read_csv('test.csv', header=None)
df2
```

```python
# index_col 옵션은 기본값 None, 이 때, 새로운 정수 0,1,2,3 ... 을 생성
# ex) index_col = 'C', C열을 인덱스로 불러서 사용하겠다는 선언

df3 = pd.read_csv('test.csv', index_col='C')
df3
```

```python
# names = ['열이름1', '열이름2', ...] 옵션을 통해 열 이름을 따로 설정할 수 있음
df4 = pd.read_csv('test.csv', names=['1열', '2열', '3열', '4열'])
df4
```

# 1-3 read_csv 의 쓸만한 옵션들?   

Pandas 공식 홈페이지 가면 정리된 문서 있음   
   
- sep(',) : 자료의 구분자 설정
- header('infer') : 첫 행을 열 이름으로 사용할 것인지 결정   
- names : 열의 이름을 리스트로 입력해 줄 수 있음   
- index_col(None) : 특정한 열 이름으로 인덱스 지정   
- dtype : 개별 열 or 모든 열의 자료형 지정   
- na_values : 결측값(NaN)으로 취급할 문자 지정   
- keep_default_na(True) : 결측값을 포함할지 여부 결정   
- na_filter(True) : 결측값 탐지   
- nrow : 읽어올 파일 행(Row)의 수 리턴
