# Series   
객체를 담을수 있는 1차원 자료구조

```python
#시리즈 생성
#dict 형태로 만들기
dic = {'a':1, 'b':2, 'c':3, 'd':4}

series_dic = pd.Series(dic)
series_dic
#dict의 key가 index로, value가 값으로 들어감
```

```python
# 인덱스와 값은 아래처럼 확인
print(series_dic.index)
print(series_dic.values)
```

```python
# 직접 만들기도 가능
# List 기반으로도 생성 가능

obj = pd.Series([4,5,-4,-5])
obj

print(obj.index)
print(obj.values)

# RangeIndex(start=0, stop=4, step=1)
# Series를 생성할 때, 따로 인덱스를 입력하지 않았기 때문에
# 0부터 시작하는 인덱스가 부여(start = 0)
# 4개의 인덱스 (stop = 4) / 1씩 건너서 (step = 1)
```

```python
#인덱스 지정 가능
obj2 = pd.Series([4,5,-4,-5], index = ['a','b','c','d'])
obj2
```

```python
# 튜플로 만들기
tup_data = ('Beaver','2024-08-01', '남', 'True')
series_data = pd.Series(tup_data, index = ['name','birth','gender','cute'])
series_data
```

```python
# 요소 선택 / 인덱싱(indexing)
print(series_data[0])
print(series_data['name'])

print(series_data.loc['name'])
print(series_data.iloc[2])
```

```python
# 여러 값을 인덱싱할 때에는 위 경우처럼 값만 반환하지 않고,
# 짝을 이루는 인덱스와 값을 모두 반환함

#주의점
# 각각의 위치, 이름으로 인덱싱 할 때에는 대괄호를 2개 씌워줘야함

print(series_data[[1,2]])
print()
print(series_data[['birth','gender']])

print(series_data.loc[['birth','gender']])
```

```python
# : 로 인덱싱 할 때는 대괄호 하나만 써도 무방
print(series_data[1:3])
print()
print(series_data['birth':'gender'])
#인덱스의 이름으로 사용하면 범위 끝('gender')이 포함됨 (정수 인덱싱은 -1이었음)
```

```python
#Boolean 배열을 사용해서 값 걸러내기 or 산술 연산 or 수학 함수 적용 O
obj2 = pd.Series([4,5,-4,-5], index = ['a','b','c','d'])
print(obj2[obj2>0])
print()
print(obj2*2)
print()

#요소 검색도 가능
print('b' in obj2)
```

```python
ticket = {'busan': 30000, 'jeju': 50000, 'jeonju': 40000, 'daejeon': 55000}
ticketS = pd.Series(ticket)
ticketS
```

```python
city = ['seoul','busan', 'jeju', 'jeonju']
a = pd.Series(ticket, index=city)
a
```

```python
# 누락된 값 찾을 때
# Nan 이면 True 가 반환됨
pd.isnull(a)
```

```python
# 반대로 누락되지 않은 것 찾을 때
pd.notnull(a)
a.notnull()
```

# 2. 데이터 프레임(Data Frame)   
데이터프레임은 시리즈가 여러개 합쳐진 자료형   
표, 엑셀과 같은 스프레드시트 형식의 자료구조   
다양한 컬럼이 있고, 각각의 컬럼안에는 서로 다른 종류의 자료형   
(숫자, 문자열, 논리 등등 ...) 이 담김   
   
DataFrame : row(행), column(열)에 대한 index를 가진 데이터 객체

```python
# dict 로 dataFrame 생성

dict_data = {'c0':[1,2,3], 'c1':[4,5,6], 'c2':[7,8,9], 'c3':[10,11,12]}
df = pd.DataFrame(dict_data)
print(type(df))
df

# 1개의 열들이 각각 Series에 해당함
# dict의 key 값이 해당 열의 이름이 됨
```

```python
# 리스트로 DF 생성
# 같은 길이의 리스트에 담긴 dict or numpy를 이용

data = {  'time':['morning','afternoon','night'],
          'startHour':[6,12,18],
          'behavior': ['getup','lunch','gohome']
        }
data1 = pd.DataFrame(data)
data1

```

```python
# columns 옵션을 선언하면, 해당 순서대로 df 객체가 생성됨
pd.DataFrame(data, columns = ['time','behavior','startHour'])
```

```python
# 리스트로 만들기 2
df = pd.DataFrame([ [21, '남', '00대'], [22, '여', 'ㅁㅁ대']],
                    index = ['김비버','박비버'],
                    columns = ['age','gender','school']
                  )
print(df)
print(df.index)
print(df.columns)
```

```python
# 행, 열의 이름 변경 : 기본적인 방식
df.index = ['student1','student2']
df.columns = ['나이','성별','학교']
df

# 인덱스와 열의 이름의 수에 맞춰서 선언해줘야함
```

```python
# 행, 열의 이름 변경 : rename() 함수
# df.rename(column = {변경 전 이름 : 변경 후 이름, ... 반복}, index={변경 전 이름 : 변경 후 이름, ... 반복}, inplace=True)
# inplace -> df에 선언한 값들을 바로 적용하겠다는 선언

df = pd.DataFrame([ [21, '남', '00대'], [22, '여', 'ㅁㅁ대']],
                    index = ['김비버','박비버'],
                    columns = ['age','gender','school']
                  )
```

```python
# 열 이름 변경
df.rename(columns = {'age':'연령','gender':'성별','school':'소속'}, inplace=True)
df
```

```python
# 행 이름 변경
df.rename(index = {'김비버':'김씨', '박비버':'박씨'}, inplace=True)
df
```

# 행을 삭제하는 drop() 함수

```python
exam_data = {'math':[90,80,70], 'korean':[98,89,95], 'english':[85,95,100], 'history':[100,90,90]
             }
df = pd.DataFrame(exam_data, index = ['student1','student2','student3'])
df
```

```python
# drop('지울 행 or 열의 이름', axis = 0 or 1, inplace=True)

# Numpy :
# axis 0 -> 열, axis = 1 ->  행 (???)
# sum으로 압축하는 형태 설명하신건가?

# Pandas : 반대로 되어있음
# axis 0 -> 행, axis = 1 -> 열

# 행을 삭제하려면 axis = 0 (default)
# 열을 삭제하려면 axis = 1

df2 = df.copy()
df2.drop('student2', axis=0,inplace=True)
df2
```

```python
# 열 삭제
df3 = df.copy()
df3.drop('math', axis=1, inplace=True)
df3
```

# DataFrame indexing : loc, iloc   
   
loc, iloc은 가장 많이 사용되는 함수   
데이터 전처리 할 때 인덱싱이 가장 많이 사용되기 때문   
loc : 인덱스의 이름을 키값으로 사용   
iloc : 인덱스의 번호를 키값으로 사용   


```python
city = pd.DataFrame({'Korea': ['Seoul', 'Jeju', 'Bucheon'],
                     'US': ['LA','Califonia','Boston'],
                     'UK': ['London','LiverPool','Cambridge'],
                     'China': ['Shanhai', 'Bejing', 'Chintao']},
                     index = ['A','B','C']
                     )
city
```

```python
first = city.loc['A']
print(first)
print()
second = city.iloc[1]
print(second)
```

```python
exam_data = {'math':[90,80,70], 'korean':[98,89,95], 'english':[85,95,100], 'history':[100,90,90]
             }
df = pd.DataFrame(exam_data, index = ['kim','lee','park'])
df
```

```python
# kim 과 lee의 점수 출력

# 두 개 이상의 행을 선택할 때는 대괄호를 신경쓰기
# 각각 선택은 대괄호 2개, 범위 선택시( : )에는 1개
print(df.loc[['kim','lee']])
print()
print(df.iloc[0:2])
print()
print(df.iloc[[0,1]])
print()
```

```python
# 열 선택
# 따로 함수가 필요 없음
# df.열이름, df['열이름'] 으로 원래 하던대로 값 호출이 가능
print(df.math)
print()
print(df['math'])
```

```python
# 범위 슬라이싱
exam_data = {'name':['Oh','Woo','Ah','Ha','Kim','Lee','Park'],
             'math':[90,80,70,60,50,40,30],
             'eng':[98,89,95,85,45,76,77],
             'kor':[85,95,100,56,25,77,98],
             'pt':[100,90,90,80,80,70,70]
             }
df = pd.DataFrame(exam_data)
df
```

```python
# 0행부터 5행까지 2행 간격으로 출력
b = df.iloc[0:6:2]

print(b)
```

```python
# 전체 데이터 역순 정렬 후 모든 행을 -1 행 간격으로 출력
c = df.sort_index(ascending=False)
print(c)
print()
c_1 = c.iloc[::-1]
print(c_1)
```
