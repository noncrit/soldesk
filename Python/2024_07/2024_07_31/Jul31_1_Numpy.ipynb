# Numpy 함수   
# 하나의 array에 적용(단항 유니버셜 함수)
유니버셜 함수 :   
ndarray 안에 있는 데이터의 요소별로 연산을 수행해주는 함수   
  

```python
import numpy as np

ar1 = np.random.randn(5,3)
ar1
```

```python
# 각 요소들의 절대값
np.abs(ar1)
```

```python
# 각 요소들의 제곱근
np.sqrt(ar1)
```

```python
# 각 요소의 제곱 계산
np.square(ar1)
```

```python
# 각 로그를 자연로그, 상용로그, 밑이 2인 로그를 씌운 값 계산
#자연 로그
print(np.log(ar1))
print()
#상용 로그
print(np.log10(ar1))
print()
#밑이 2인 로그
print(np.log2(ar1))
```

```python
# 각 요소의 부호 나타내기 ( + : 1, - : -1, 0 : 0)
np.sign(ar1)
```

```python
# 각 요소의 올림 / 내림 / 반올림
r = np.ceil(ar1)
r1 = np.floor(ar1)
r2 = np.round(ar1)
print(r)
print(r1)
print(r2)
```

```python
# 각 요소가 nan인 경우 True, 아니면 False를 반환
np.isnan(ar1)
```

```python
# 각 요소가 무한대인 경우 True, 아닌 경우 False
np.isinf(ar1)
```

```python
# 각 요소에 대해 삼각함수 값을 계산하기 (cos, cosh, sin, sinh, tan, tanh)
# -h : 쌍곡선 함수
np.cos(ar1)
```

# 두 개의 array에 적용 가능한 함수 ( 이항 유니버셜 함수 )

```python
ar2 = np.random.randn(5,3)
ar2
```

```python
# 두 개의 array에 대해 동일한 위치 요소끼리 연산
# 계산함수(add, substract, multiply, divide)

np.multiply(ar1, ar2)
```

```python
# 두 개의 array에 대해 동일한 위치 요소끼리 비교해서 최대값 or 최소값 계산
# maximum, minimum
np.maximum(ar1, ar2)
```

# 통계 함수   
통계 함수를 통해서 array의 합이나 평균들을 구할 때   
추가로 axis 라는 파라미터에 대한 값을 지정해서   
열 또는 행의 합, 평균, 외 기타 등등에 대한 값을 구할 수 있음

```python
# 열 간의 합을 계산
# Numpy에서 axis = 0 은 열을 의미함
# axis = 1 은 행을 의미함
np.sum(ar1, axis=0)
```

```python
np.sum(ar1, axis=1)
```

```python
# 전체 요소의 평균 계산
np.min(ar1, axis=0)
```

```python
# 전체 요소의 표준 편차, 분산, 최소값, 최대값
# 표준편차 : 데이터가 평균으로부터 얼마나 흩어져있는지를 나타내는 지표
# 분산 : 데이터가 평균으로부터 얼마나 멀리 있는지를 나타내는 지표
# std, var, min, max
np.std(ar1)
```

```python
# 전체 요소 중 최소값, 최대값이 위치한 인덱스 반환
# argmin, argmax
print(ar1)
np.argmin(ar1)
```

```python
# 맨 처음 요소부터 각 요소까지의 누적합 or 누적곱을 계산 (cumsum, cumprod)
np.cumsum(ar1)
```

```python
# 사분위수
# 데이터를 4등분 했을 때 0%, 25%, 50%, 100% 에 해당하는 값들
np.quantile(ar1, [0, 0.25, 0.5, 0.75])
```

# 기타 함수

```python
# 전체 요소에 대해 오름차순 정렬
np.sort(ar1)
```

```python
# 행 / 열 방향 지정해서 정렬

# 열 방향으로 정렬
np.sort(ar1, axis=0)
```

```python
# zip 함수
# 각 객체들을 파라미터로 받고, 그 요소들을 tuple의 형태로 반환
number = [1,2,3]
letters = ['A', 'B', 'C']
for pair in zip(number, letters):
    print(pair)
```
