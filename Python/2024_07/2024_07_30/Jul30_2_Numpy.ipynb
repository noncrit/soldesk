#Part 2. 배열   

Numpy의 배열은 모두 같은 자료형을 사용하는 값의 집합이고,   
음수가 아닌 정수의 튜플로 인덱싱   
   
배열의 shape는 각 차원별 크기의 tuple   
중첩된 파이썬 리스트로부터 numpy 배열을 초기화   
대괄호를 사용해서 각각의 요소로 접근   
   


```python
#ex
ex = [ [1,2,3], [4,5,6], [7,8,9] ]
print(ex)
print(np.array(ex))
exx = np.array(ex)

#shape를 통해 몇 개의 데이터가 있는지, 몇 차원의 데이터인지 알 수 있음
exx.shape
```

```python
a = np.array([1,2,3])
print(a)

#type - 자료형
#ndim -> n -dimension -> 몇 차원인지
#shape -> 튜플 형식으로 배열의 크기, 차원을 표현 (배열 크기, 차원) 현재는 1차원이라 튜플 형태만 유지함
#dtype -> 배열 요소들의 자료형 (int64는 자바로 따지면 long 정도)

print(type(a), a.ndim, a.shape, a.dtype)
#<class 'numpy.ndarray'> 1 (3,) int64
```

# Array 연산   
   
기본적으로 Numpy에서 연산을 하는 경우, 서로 크기가 같은 array 끼리 연산을 진행   
같은 위치에 있는 요소들끼리 연산을 진행   


```python
arr1 = np.array([ [1,2,3], [4,5,6]])
arr1
```

```python
arr2 = np.array( [ [10,11,12], [13,14,15] ] )
arr2
```

# Indexing 과 Slicing의 차이   

   indexing을 사용하게 되면, 항상 랭크(차원)이 감소,   
   slicing을 사용하면 랭크(차원)이 유지됨

```python
#몇 차원인가? => 중괄호 갯수로 파악하면 편함
b = np.arange(1,13).reshape(3,2,2)
print(b)
print(b.ndim)
```

```python
#인덱싱 -> 차원 감소  / 슬라이싱 -> 차원 유지
indexing = b[1]
slicing =  b[1:2]
print(indexing, indexing.shape, indexing.ndim)
print(slicing, slicing.shape, slicing.ndim)
```

```python
#array 수정

arr = np.array([11,22,33,44,55])
print(arr)

# 0 번째 인덱스의 값을 10으로 변경
arr[0] = 10
print(arr[0])

# 1,2 번째 인덱스에 있는 값을 222,333 으로 변경
arr[1:3] = np.array([222,333])
print(arr)

# 3,4 번째 인덱스에 있는 값에 2를 더하기
arr[3:5] += 2
print(arr)
```

```python
arr = np.array([110,222,337,420,155,987,200,53,178])
print(arr)

# arr 에서 200보다 큰 요소들에 대해서 기존 값에 2를 더한 값으로 수정
# call by ref 방식이라서 인덱스에 직접 접근해서 값을 수정해야함, 안그러면 원래 주소 참조해서 값이 안바뀜...
for k,v in enumerate(arr):
  if v>200:
    arr[k] = v+2
print(arr)
```

```python
arr = np.array([110,222,337,420,155,987,200,53,178])
print(arr)

arr[arr>200] += 2
print(arr)
```

```python
# 배열 요소 삭제

arr = np.array([11,22,33,44,55])
print(arr)

#0번째, 2번째 인덱스의 값을 삭제하고 싶을때
# np.delete(target_array, index)
arr = np.delete(arr, [0, 2])
print(arr)
```

```python
# array 인덱스 추출
# np.where(조건)

# arr의 각 요소의 값이 100보다 작은 index 추출
np.where(arr<100)
```

```python
arr = np.array([11,130,370,114,55,94,200,30,190,187])

#arr의 각 요소의 값이 100이상이면 'A' 출력, 아니라면 'B' 출력
result = 'A'
for i in arr:
  if i<100:
    result = 'B'
print(result)

#원래 의도한 출력은 이 구현
#np.where 사용하는 버전
result2 = np.where(arr>=100, 'A','B')
print(result2)
```
